```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

results = [
    dict(model='alexnet_baseline', normalized_score=0.38632657),
    dict(model='alexnet_lrm_pass0', normalized_score=0.38963921),
    dict(model='alexnet_lrm_pass1', normalized_score=0.93958962),
    dict(model='alexnet_lrm_pass2', normalized_score=0.88334215),
]
df = pd.DataFrame(results)
df
```

```python
benchmark_identifier
```

```python
# Set the figure size using figsize (width, height) in inches
plt.figure(figsize=(8, 6))  # Create a figure that's 8 inches wide and 6 inches tall

benchmark_identifier=f'Marques2020_Cavanaugh2002-{property_name}'
lrm_models = ['alexnet_lrm_pass0', 'alexnet_lrm_pass1', 'alexnet_lrm_pass2']
ax = sns.barplot(data=df[df.model.isin(['alexnet_baseline'])], x="model", y="normalized_score", color="orange", label="pytorch")
ax = sns.barplot(data=df[df.model.isin(lrm_models)], x="model", y="normalized_score", color="blue", label="lrm")
ax.set_ylim([0,1.1])
ax.set_title(benchmark_identifier, fontsize=16, pad=10)
plt.axhline(y=1.0, color='grey', linestyle='--')

# Add "noise ceiling" in italics, centered above the line using data coordinates for both x and y
x_center = 0.5 * (ax.get_xlim()[0] + ax.get_xlim()[1])  # Calculate the midpoint of the x-axis
plt.text(x=x_center, y=1.00 + 0.02, s=r'$\it{noise\ ceiling}$', color='gray', fontsize=12, style='italic',
         horizontalalignment='center')

plt.xlabel("model", fontsize=22, labelpad=10)
plt.ylabel("brain-score\n(noise-adjusted)", fontsize=22, labelpad=10)

# Display the plot
plt.show()
```
